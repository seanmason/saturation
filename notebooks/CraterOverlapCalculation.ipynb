### Crater Overlap Calculation

The following plots demonstrate crater overlap calculation. In these calculations, an existing crater is overlapped by a new, incoming crater. The existing crater is shown in green, the new crater in black, and the crater rim section to be destroyed in blue.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from saturation.simulation import *
from saturation.geometry import *
from saturation.plotting import *
```

```python
def plot_intersection(center1: Location, 
                      radius1: float,
                      center2: Location,
                      radius2: float,
                      scale: float):
    fig, ax = plt.subplots(figsize=(5, 5))

    ax.set_xlim([0, scale])
    ax.set_ylim([0, scale])

    # Plot craters
    plot_circle(center1, radius1, ax, color='green')
    plot_circle(center2, radius2, ax)

    # Plot intersection
    arc = get_intersection_arc(center1, radius1, center2, radius2)
    plot_arc(center1, radius1, arc, ax, color='blue', lw=3)

    plt.show()
```

```python
plot_intersection((10, 10),
                  10,
                  (20, 10),
                  10,
                  30)
```

```python
plot_intersection((20, 10),
                  10,
                  (10, 10),
                  10,
                  30)
```
