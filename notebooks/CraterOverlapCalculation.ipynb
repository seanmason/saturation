### Crater Overlap Calculation

The following plots demonstrate crater overlap calculation. In these calculations, an existing crater is overlapped by a new, incoming crater. The existing crater is shown in green, the new crater in black, and the crater rim section to be destroyed in blue.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from saturation.simulation import *
from saturation.geometry import *
```

```python
def plot_crater(center: Location, radius: float, arc: Arc, size: float, axes_subplot, color: str = 'black', lw: float = 1):
    axes_subplot.add_patch(matplotlib.patches.Arc(center,
                                                  width=radius * 2,
                                                  height=radius * 2,
                                                  theta1=np.rad2deg(arc[0]),
                                                  theta2=np.rad2deg(arc[1]),
                                                  color=color,
                                                  lw=lw))

def plot_intersection(center1: Location, center2: Location, radius1: float, radius2: float, size: float):
    fig, ax = plt.subplots(figsize=(5, 5))

    ax.set_xlim([0, size])
    ax.set_ylim([0, size])

    # Plot craters
    plot_crater(center1, radius1, (0, 2*np.pi), size, ax, color='green')
    plot_crater(center2, radius2, (0, 2*np.pi), size, ax)

    # Plot intersection
    arc = get_intersection_arc(center1, center2, radius1, radius2)
    plot_crater(center1, radius1, arc, size, ax, color='blue', lw=3)

    plt.show()
```

```python
plot_intersection((10, 10),
                  (20, 10),
                  10,
                  10,
                  30)
```
