```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from saturation.simulation import *
from saturation.distributions import *
```

### Drawing Craters

```python
power_law_distribution = PowerLawProbabilityDistribution(slope=-2.8, min_value=8)
craters = get_craters(500, power_law_distribution)
```

```python
size = 1000

fig, ax = plt.subplots(figsize=(8, 8))

ax.set_xlim([0, size])
ax.set_ylim([0, size])

for crater in craters.itertuples():
    center = (crater.x * size, crater.y * size)
    ax.add_patch(matplotlib.patches.Arc(center,
                                        width=crater.radius,
                                        height=crater.radius,
                                        theta1=0,
                                        theta2=360))
    
plt.show()
```
