### Drawing Craters

```python
import numpy as np
import itertools
import matplotlib
import matplotlib.pyplot as plt

from saturation.simulation import *
from saturation.distributions import *
from saturation.geometry import *
from saturation.plotting import *
```

```python
observed_terrain_size = 1000
dist = ParetoProbabilityDistribution(cdf_slope=1.8, x_min=8, x_max=500)
craters = list(itertools.islice(get_craters(dist, observed_terrain_size), 500))
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.set_xlim([0, observed_terrain_size])
ax.set_ylim([0, observed_terrain_size])

for crater in craters:
    plot_circle((crater.x, crater.y), crater.radius, ax)
    
plt.show()
```
