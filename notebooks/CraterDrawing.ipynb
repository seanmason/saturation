```python
import numpy as np
import itertools
import matplotlib
import matplotlib.pyplot as plt

from saturation.simulation import *
from saturation.distributions import *
```

### Drawing Craters

```python
power_law_distribution = PowerLawProbabilityDistribution(slope=-2.8, min_value=8)
craters = list(itertools.islice(get_craters(power_law_distribution), 500))
```

```python
size = 1000

fig, ax = plt.subplots(figsize=(8, 8))

ax.set_xlim([0, size])
ax.set_ylim([0, size])

for crater in craters:
    center = (crater.location[0] * size, crater.location[1] * size)

    for rim_segment in crater.rim_segments:
        ax.add_patch(matplotlib.patches.Arc(center,
                                            width=crater.radius,
                                            height=crater.radius,
                                            theta1=np.rad2deg(rim_segment[0]),
                                            theta2=np.rad2deg(rim_segment[1]))) 
plt.show()
```
